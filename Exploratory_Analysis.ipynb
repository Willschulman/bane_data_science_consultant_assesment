Clean Date Fields

```python
##cleaning up date fields

df['FOUNDED_ON'] = pd.to_datetime(df['FOUNDED_ON'], errors='coerce', format='%Y-%m-%d')
df['LAST_FUNDING_ON'] = pd.to_datetime(df['LAST_FUNDING_ON'], errors='coerce', format='%Y-%m-%d')


df['founded_year'] = df['FOUNDED_ON'].dt.year
df['founded_month'] = df['FOUNDED_ON'].dt.month
df['founded_half'] = df['founded_month'].apply(lambda x: 1 if x <= 6 else 2)

df['lfunded_year'] = df['LAST_FUNDING_ON'].dt.year
df['lfunded_month'] = df['LAST_FUNDING_ON'].dt.month
df['lfunded_half'] = df['lfunded_month'].apply(lambda x: 1 if x <= 6 else 2)
```

Explode Categories so that individual categories can be analyzed 

```python

cat_df = df
cat_df['categories'] = cat_df['CATEGORY_GROUPS_LIST'].fillna('').str.split(',')
cat_df['weight'] = cat_df['TOTAL_FUNDING_USD'] / cat_df['categories'].apply(len)
cat_df = cat_df.explode('categories')
cat_df = cat_df[cat_df['categories']!= 'Other']

```

```python
weighted_top_10 = cat_df.groupby('categories').agg({'weight':'sum'}).sort_values(by='weight', ascending=False).head(10)
unweighted_top_10 = cat_df.groupby('categories').agg({'TOTAL_FUNDING_USD':'sum'}).sort_values(by='TOTAL_FUNDING_USD', ascending=False).head(10)
funding_per_company = cat_df.groupby('categories').agg({'UUID':pd.Series.nunique,'TOTAL_FUNDING_USD':sum})
funding_per_company['funding_per_company'] = (funding_per_company['TOTAL_FUNDING_USD'] / funding_per_company['UUID']).astype(int)
funding_per_company_top_10 = funding_per_company.sort_values(by='funding_per_company', ascending=False).head(10)
```

```python

fig, ax = plt.subplots(1, 3, figsize=(20, 6))

# Weighted Top 10
ax[0].bar(weighted_top_10.index, weighted_top_10['weight'], color='b')
ax[0].set_title('Top 10 Categories by Weighted Sum')
ax[0].tick_params(axis='x', rotation=90)
ax[0].set_ylabel('Weighted Sum')

# Unweighted Top 10
ax[1].bar(unweighted_top_10.index, unweighted_top_10['TOTAL_FUNDING_USD'], color='g')
ax[1].set_title('Top 10 Categories by Unweighted Sum')
ax[1].tick_params(axis='x', rotation=90)
ax[1].set_ylabel('Unweighted Sum')


# Funding Per Company Top 10
ax[2].bar(funding_per_company_top_10.index, funding_per_company_top_10['funding_per_company'], color='r')
ax[2].set_title('Top 10 Categories by Funding per Company')
ax[2].tick_params(axis='x', rotation=90)
ax[2].set_ylabel('Funding per Company')

plt.tight_layout()
plt.show()
```

Change in category funding over time

```python
weighted_year = cat_df.groupby(['categories','lfunded_year']).agg({'weight':'sum','RAISED_AMOUNT_USD':'sum'}).sort_values(by='weight', ascending=False).reset_index()
weighted_year = weighted_year[weighted_year['lfunded_year']>2009]
weighted_year = weighted_year[weighted_year['categories']!= '']
```

```python
# Calculate the total funding per year
total_funding_per_year = weighted_year.groupby('lfunded_year')['weight'].sum()

weighted_year['proportion'] = weighted_year.apply(lambda row: row['weight'] / total_funding_per_year[row['lfunded_year']], axis=1)

weighted_year = weighted_year.sort_values('lfunded_year')

top_categories_per_year = weighted_year.groupby('lfunded_year').apply(lambda x: x.nlargest(10, 'proportion')).reset_index(drop=True)

filtered = weighted_year.merge(top_categories_per_year, on=['lfunded_year', 'categories', 'proportion'])

plt.figure(figsize=(15, 8))
for category in filtered['categories'].unique():
    subset = filtered[filtered['categories'] == category]
    plt.plot(subset['lfunded_year'], subset['proportion'], label=category, marker='o')

plt.title('Year-Specific Top 5 Categories by Proportion of Total Funding Over Time')
plt.ylabel('Proportion of Total Funding')
plt.xlabel('lfunded_year')
plt.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

Look at regional differences in categories

```python
# Define the region mapping
region_mapping = {
    'Northeast': ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 
                  'Connecticut', 'New York', 'New Jersey', 'Pennsylvania'],
    'Midwest': ['Ohio', 'Indiana', 'Michigan', 'Illinois', 'Missouri', 'Wisconsin', 'Minnesota', 
                'Iowa', 'Kansas', 'Nebraska', 'South Dakota', 'North Dakota'],
    'South': ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 
              'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 
              'Oklahoma', 'Texas', 'Arkansas', 'Louisiana'],
    'West': ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 
             'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
}

# Map states to regions
def map_to_region(state):
    for region, states in region_mapping.items():
        if state in states:
            return region
    return 'Other'

cat_df['REGION_GROUP'] = cat_df['REGION'].apply(map_to_region)

grouped = cat_df.groupby(['categories', 'REGION_GROUP'])['weight'].sum().reset_index()
top_categories = grouped.groupby('categories')['weight'].sum().nlargest(10).index
filtered = grouped[grouped['categories'].isin(top_categories)]

filtered = filtered.sort_values(by=['REGION_GROUP', 'weight'], ascending=[True, False])

plt.figure(figsize=(15, 8))
ax = sns.barplot(data=filtered, y='weight', x='REGION_GROUP', hue='categories', ci=None)

plt.title('Top 10 Categories by Weighted Funding Split by REGION_GROUP')
plt.ylabel('Weighted Funding')
plt.xlabel('REGION_GROUP')
plt.xticks(rotation=45)
plt.legend(title='Categories', loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
from datetime import datetime

```

Calculate number of startups founded in h1 2022 vs h1 2023

Look into change in Geographic concentrations in number of startups

```python
top_cities['difference'] = top_cities['2023%'] - top_cities['2022%']

labels = top_cities['CITY'].values
difference = top_cities['difference'].values
top_city_values = top_cities['CITY'][top_cities['difference'] >= 0.01].values


x = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(12, 8))

colors = []
for diff in difference:
    if diff >= 0.01:
        colors.append('darkgreen')
    elif diff >= 0:
        colors.append('lightgreen')
    else:
        colors.append('red')


bars = ax.bar(x, difference, color=colors)

ax.set_ylabel('Difference in Percentage')
ax.set_title('Difference in Geographic Concentrations of Founded Startups from H1 2022 to H1 2023')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=90, ha='right')
ax.axhline(0, color='black', linewidth=0.8)  

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3 if height > 0 else -3),  
                    textcoords="offset points",
                    ha='center', va='center', fontsize=8)

autolabel(bars)

plt.tight_layout()
plt.show()

```

Begin Data transformations for Investors investigation

```python
inv_df = df
inv_df['categories'] = inv_df['CATEGORY_GROUPS_LIST'].fillna('').str.split(',')
inv_df['investors'] = inv_df['LEAD_INVESTOR_UUIDS'].fillna('').str.split(',')
inv_df['weight'] = inv_df['RAISED_AMOUNT_USD'] / inv_df['categories'].apply(len) / inv_df['investors'].apply(len)

inv_df = inv_df.explode('categories')
inv_df = inv_df.explode('investors')

investors_raised = inv_df.groupby('investors').agg({'weight': 'sum'}).sort_values(by='weight', ascending=False).reset_index()
investors_raised = investors_raised[investors_raised['investors']!='']

top_5 = investors_raised[investors_raised['weight'] >= investors_raised['weight'].quantile(0.95)]

top_5 = top_5['investors']

top_inv_df = inv_df[inv_df['investors'].isin(top_5)]

top_inv_df = top_inv_df.groupby(['categories'])['weight'].sum().reset_index().sort_values(by='weight', ascending=False)
```

Plot Investor category breakdown

```python
rest_inv_df = inv_df[~inv_df['investors'].isin(top_5)]
rest_agg = rest_inv_df.groupby('categories')['weight'].sum().reset_index()

merged_df = top_inv_df.merge(rest_agg, on='categories', how='outer', suffixes=('_top_5', '_rest'))

merged_df.fillna(0, inplace=True)
merged_df.reset_index(drop=True, inplace=True)
# Exclude the 'Other' category
merged_df = merged_df[merged_df['categories'] != 'Other']

merged_df['combined_weight'] = merged_df['weight_top_5'] + merged_df['weight_rest']
merged_df = merged_df.sort_values(by='combined_weight', ascending=False).head(20)

plt.figure(figsize=(15, 10))
bar_top_5 = sns.barplot(data=merged_df, x='categories', y='weight_top_5', label='Top 5% Investors', color='blue')
bar_rest = sns.barplot(data=merged_df, x='categories', y='weight_rest', label='Rest of Investors', color='red', bottom=merged_df['weight_top_5'])

for idx in range(len(merged_df)):
    total = merged_df['weight_top_5'].iloc[idx] + merged_df['weight_rest'].iloc[idx]
    percentage = (merged_df['weight_top_5'].iloc[idx] / total) * 100
    rect = bar_top_5.patches[idx]
    x = rect.get_x() + rect.get_width() / 2
    y = rect.get_y() + rect.get_height() + 0.02 * total  # A little above the top of the bar
    plt.text(x, y, f"{percentage:.0f}%", ha='center', va='center')

plt.legend(loc="upper right")
plt.xticks(rotation=90)
plt.ylabel('Investment Weight')
plt.title('Comparison of Investment Categories between Top 5% and Rest of the Investors (Top 20 Categories)')
plt.tight_layout()  
plt.show()
```
